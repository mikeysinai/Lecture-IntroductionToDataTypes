### ‚ö° Extra: How Booleans Work in Python


Almost everything in Python converts to `True` when turned into a Boolean:  

- **False values**: `0`, `0.0`, `""` (empty string), `[]` (empty list), `{}` (empty dict), and `None`.  
- **Everything else** is `True`.  


```python

print(bool(0))         # False
print(bool(5))         # True
print(bool(""))        # False
print(bool("Hello"))   # True
print(bool([]))        # False
print(bool([1, 2]))    # True
```

*Boolean values in Python vs other programming languages*.

You may have noticed that the data we grabbed from the web used true/false instead of True/False.  In Python you must use the capital letters.
